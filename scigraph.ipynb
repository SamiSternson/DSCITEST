```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
UV=pd.read_csv("UV1_5DAYWater.csv")
NUV=pd.read_csv("NUV1_5DAYWater.csv")
print(UV.loc[11])
```

```python

points=np.array(UV.loc[11])
print(points)
NUVpoints=np.array(NUV.loc[11])
print(NUVpoints)
```

```python
points=points[~np.isnan(points)]
print(points)
NUVpoints=NUVpoints[~np.isnan(NUVpoints)]
print(NUVpoints)
```

```python
uv_list=[]
nuv_list=[]
for i in range(len(points)):
    uv_list.append("EXPOSED TO UV")
for i in range(len(NUVpoints)):
    nuv_list.append("NOT EXPOSED TO UV")
ax, fig=plt.subplots()
plt.rcParams['font.size'] = 18
plt.bar(uv_list, np.mean(points),color="blue", yerr=np.std(points),ecolor='black', capsize=10)
plt.bar(nuv_list, np.mean(NUVpoints),color="blue", yerr=np.std(NUVpoints),ecolor='black', capsize=10)
plt.scatter(uv_list, points, c="red")
plt.scatter(nuv_list, NUVpoints, c="red")
plt.show()
```
